```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import h5py
import yaml
import os.path
```

```python
# change this directory to the one where the data is stored
data_dir = '/Users/jpga/Documents/reconstruction_scan'
```

```python
# test file name
test_file_name = 'DYG14_1701987709.h5'
```

```python
# open h5 file as a Python object
test_file = h5py.File(os.path.join(data_dir, test_file_name), 'r')
```

```python
# show h5 file keys
test_file.keys()
```

```python
# store images in a numpy array and show its shape. The first dimension is the number of images, and the last two are the 2D image dimensions
test_images = test_file['images'][:]
test_images.shape
```

```python
# plot the first image:
plt.imshow(test_images[0])
```

```python
# show attributes of the images (this is the metadata of the images). test_file['images'].attrs is a Python dictionary with the following keys: 
test_file['images'].attrs.keys()
```

```python
# we are interested in "AWA:Bira3Ctrl:Ch10", which is the channel that indicates the quadrupole strength for the scan.
# we can access the value of this channel as you will do with Python dictionaries:
quad_pv_name = "AWA:Bira3Ctrl:Ch10"
quad_pv = test_file['images'].attrs[quad_pv_name]
quad_pv
```
